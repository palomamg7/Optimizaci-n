```python
f=x**3-3*x**2-24*x+32
df=f.diff()
ddf=df.diff()
df,ddf
```

```python
pc=solve(df)
pc
```

```python
#PROBLEMA 1
A=x*y
eq1=Eq(2*x+y,2400)
A,eq1
```

```python
y_=solve(eq1,y)
y_
```

```python
f=A.subs(y,y_[0])
f
```

```python
df=f.diff()
ddf=df.diff()
df,ddf
```

```python
pc=solve(df)
pc
```
