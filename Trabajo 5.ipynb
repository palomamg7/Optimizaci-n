```python
%matplotlib notebook
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress 
init_printing ()
```

```python
x=np.array([-1,0,1,2])
y=np.array([0,1,2,1])
res=linregress(x,y)

xn=np.arange(-1,2,.5)
yn=res.intercept+res.slope*xn

plt.plot(x,y,'*')
plt.plot(xn,yn)
plt.plot(x,y,'*b')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
x=np.linspace(0,1,30)
y=10*x+2+np.random.randn(30)

res=linregress(x,y)
xn=np.arange(0,1,.01)
yn=res.intercept+res.slope*xn

plt.plot(xn,yn)
plt.plot(x,y,'*b')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
# Puntos
x=np.array([1,5,20,40])
y=np.array([56.5,113,181,214.5])
res=linregress(x,y)

# Linea
xn=np.arange(1,40,.5)
yn=res.intercept+res.slope*xn

plt.plot(x,y,'*')
plt.plot(xn,yn)
plt.plot(x,y,'*b')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

```

```python
#Para hacer una curva 
from scipy.optimize import curve_fit

def acetona(x,a,b,c):
    return(a+b*x+c*(x**2))

P=np.array([1,5,20,40])
T=np.array([56.5,113,181,214.5])

(A,B,C),_=curve_fit(acetona,P,T)

print(A,B,C)

#nuevos puntos de x
nxdat=np.arange(1,40,.5)
ydatAjuste=acetona(nxdat,A,B,C)

plt.plot(P,T,"o",nxdat,ydatAjuste)
plt.legend(["Datos","Ajuste"],loc="best")
```
